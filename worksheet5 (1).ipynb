```python
X = df.drop(columns=["Writing"]).values;
y= df["Writing"].values;
def train_test_split(X, y, test_size=0.3, random_seed=42):
  np.random.seed(random_seed)
  indices = np.arange(X.shape[0])
  np.random.shuffle(indices) # Shuffle the indices
  test_split_size = int(len(X) * test_size)
  test_indices = indices[:test_split_size]
  train_indices = indices[test_split_size:]
  X_train, X_test = X[train_indices], X[test_indices]
  y_train, y_test = y[train_indices], y[test_indices]
  return X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
def cost_function(X, Y, W):
  # X = np.array(X, dtype=float)
  # y = np.array(Y, dtype=float).reshape(-1, 1)
  # W = np.array(W, dtype=float).reshape(-1, 1)
  n = len(Y)
  Y_pred = X.dot(W)
  error = Y_pred - Y
  cost = (1/(2*n))*(error ** 2).sum()
  return cost
```

TO-DO-TASK(6)

```python
def gradient_descent(X, Y, W, alpha, iterations):
  cost_history = [0] * iterations
  m = len(Y)
  W_update = W.copy()
  for iteration in range(iterations):
    Y_pred = X.dot(W)
    loss = Y_pred - Y
    dw = (1/m)*X.T.dot(loss)
    W_update = W_update - alpha * dw
    cost = cost_function(X, Y, W_update)
    cost_history[iteration] = cost
  return W_update, cost_history

```

TO-DO-TASK(8)

```python
def rmse(Y, Y_pred):
  rmse=np.sqrt(np.mean((Y-Y_pred)**2))
  return rmse
```

TO-DO-TASK(9)

```python
def r2(Y, Y_pred):
  mean_y=np.mean(Y)
  ss_tot=np.sum((Y_pred - mean_y)**2)
  ss_res=np.sum((Y-Y_pred))
  r2=1-(ss_res/ss_tot)
  return r2
```
