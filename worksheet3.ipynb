Exercise 3.2

filling the missing values usinf fillna

```python
dataframe["Age"]= dataframe["Age"].fillna(dataframe["Age"].median())
```

Highest fare paid relative to age

```python
dataframe["fare_per_year"]= dataframe["Fare"]/dataframe["Age"]
high_fare_age = dataframe.loc[dataframe["fare_per_year"] > 5]
high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)
result = high_fare_age_srt[["Name", "fare_per_year"]]
print(result)
```

adult male passenger paid highest fare relative to their class

```python
dataframe["fare_by_class"]=dataframe["Fare"]/dataframe["Pclass"]
adult_males=dataframe.loc[(dataframe["Sex"] == "male") & (dataframe["Age"]>=18)]
adult_males_srt=adult_males.sort_values(by="fare_by_class", ascending=False)
result=adult_males_srt[["Name", "Age", "fare_by_class"]]
print(result)
```

3.3

Total revenue from each passenger class

```python
total_fare = dataframe["Fare"].sum()
class1_fare = dataframe.loc[dataframe["Pclass"]==1,"Fare"].sum()#didnot keep["fare"]since we wanted a number not series
class2_fare= dataframe.loc[dataframe["Pclass"]==2,"Fare"].sum()
class3_fare= dataframe.loc[dataframe["Pclass"]==3,"Fare"].sum()
classfare_list= [class1_fare, class2_fare, class3_fare]
revenue = [i/total_fare * 100 for i in classfare_list]
print(revenue)
```

percent of total number of passengers belonging to diff age group

```python
dataframe["Age"]= dataframe["Age"].fillna(dataframe["Age"].median())
def age_check(age):
  if age <18:
    return "child";
  elif age < 65:
    return "adult";
  else:
    return "senior";

dataframe["age_group"] = dataframe["Age"].apply(age_check)
total_passengers = len(dataframe)
passengers_in_each_agegroup = dataframe.groupby("age_group")["Name"].count()
passengers_in_each_agegroup_percent = (passengers_in_each_agegroup/total_passengers)*100
print(passengers_in_each_agegroup_percent)
```
