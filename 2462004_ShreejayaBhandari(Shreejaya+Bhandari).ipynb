Problem - 1A - Single Year HDI Exploration (Latest Year: 2022)

1. Extract Latest Year:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the pandas library and give it the alias 'pd'
# Pandas is used for data manipulation and analysis
import pandas as pd

# Load the Human Development Index (HDI) dataset from the given file path
# The CSV file is read and stored as a DataFrame named 'dataset'
dataset = pd.read_csv("/content/drive/MyDrive/Concept and technologies of AI/Human_Development_Index_Dataset.csv")

dataset.drop(columns=["Unnamed: 0"], inplace = True)

# Extract all unique values from the 'year' column
# This helps to identify which years are present in the dataset
unique_years = dataset['year'].unique()

# Print the unique years available in the dataset
print(unique_years)

# Filter the dataset to include only records from the year 2022
# This creates a new DataFrame containing HDI data for 2022 only
hdi_2022_df = dataset[dataset['year'] == 2022]

# Display the first 10 rows of the 2022 HDI dataset
# This is useful for previewing the structure and values of the filtered data
hdi_2022_df.head(10)
```

2. Data Exploration:

```python
# Display the first 10 rows of the filtered 2022 HDI dataset
# This helps to preview the data values and structure
first_rows = hdi_2022_df.head(10)
print(first_rows)

# Count the number of rows in the 2022 dataset
# shape[0] returns the total number of observations (rows)
rows = hdi_2022_df.shape[0]

# Count the number of columns in the 2022 dataset
# shape[1] returns the total number of variables (columns)
columns = hdi_2022_df.shape[1]

# Print the total number of rows and columns
print("Number of rows:", rows)
print("Number of columns:", columns)

# List all column names along with their data types for the year 2022
# This is useful for understanding the structure of the dataset
# and identifying numerical and categorical variables
dataset[dataset['year'] == 2022].dtypes
```

3. Missing Values & Data Cleaning:

```python
# Check for missing (null) values in each column
print(hdi_2022_df.isna().sum())

# Check for duplicate rows in the dataset
print("Duplicates:", hdi_2022_df.duplicated().sum())

# Create a copy of the dataset to avoid modifying the original data
hdi_2022_df = hdi_2022_df.copy()

# Replace common missing value symbols (–, -, NA, empty string) with proper NaN values
hdi_2022_df.replace(['–', '-', 'NA', ''], pd.NA, inplace=True)

# Remove duplicate rows from the dataset
hdi_2022_df = hdi_2022_df.drop_duplicates()

# Confirm that the data cleaning process is completed
print("Cleaning done!")
```

4. Basic Statistics:

```python
# Computing basic descriptive statistics for the HDI column

# Calculate the mean (average) HDI value for 2022
mean_hdi = hdi_2022_df['hdi'].mean()

# Calculate the median (middle) HDI value for 2022
median_hdi = hdi_2022_df['hdi'].median()

# Calculate the standard deviation to measure HDI variability
std_hdi = hdi_2022_df['hdi'].std()

# Display the computed statistics
print("Mean HDI (2022):", mean_hdi)
print("Median HDI (2022):", median_hdi)
print("Standard Deviation of HDI (2022):", std_hdi)

# Identify the country with the highest HDI value in 2022
highest_hdi_country = hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax(), 'country']
print("Country with highest HDI in 2022:", highest_hdi_country)

# Identify the country with the lowest HDI value in 2022
lowest_hdi_country = hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin(), 'country']
print("Country with lowest HDI in 2022:", lowest_hdi_country)
```

5. Filtering and Sorting:

```python
# Filter countries with high HDI (greater than 0.800)
high_hdi = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]

# Display all countries with high HDI
print(high_hdi)

# Sort the high-HDI countries by Gross National Income per capita in descending order
high_hdi_sorted = high_hdi.sort_values(by='gross_inc_percap', ascending=False)

# Display the sorted list
print(high_hdi_sorted)

# Show the top 10 high-HDI countries with the highest income per capita
high_hdi_sorted.head(10)

```

6. Adding HDI Category Column:

```python

# Function to classify HDI values into development categories
def classify_hdi(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi <= 0.699:
        return "Medium"
    elif 0.700 <= hdi <= 0.799:
        return "High"
    else:
        return "Very High"

# Create a new column that stores the HDI category for each country
hdi_2022_df['hdi_category'] = hdi_2022_df['hdi'].apply(classify_hdi)

# Display the count of countries in each HDI category
print(hdi_2022_df['hdi_category'].value_counts())

# Check the first 10 rows with country name, HDI value, and category
hdi_2022_df[['country', 'hdi', 'hdi_category']].head(10)

# Save the final dataset with the new HDI category column as a CSV file
hdi_2022_df.to_csv("hdi_category_added.csv", index=False)

# Confirm that the file has been saved
print("File saved as: hdi_category_added.csv")
```

2. Composite Development Score:

```python
# South Asian countries
south_asia_list = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

# Filter South Asia data
sa = hdi_2020_2022[hdi_2020_2022['country'].isin(south_asia_list)].copy()

# Create Composite Score
sa['Composite_Score'] = (
    0.3 * sa['life_expectancy'] +
    0.3 * sa['gross_inc_percap']
)

# Use only 2022 data
sa_2022 = sa[sa['year'] == 2022]

# Rank countries
ranked = sa_2022.sort_values('Composite_Score', ascending=False)

# Show results
ranked[['country', 'Composite_Score', 'hdi']]

# Top 5 countries
top5 = ranked.head(5)

# Plot
plt.figure()

sns.barplot(data=top5, x='Composite_Score', y='country')

plt.title("Top 5 South Asian Countries by Composite Score (2022)")
plt.xlabel("Composite Development Score")
plt.ylabel("Country")

plt.show()
```

3. Outlier Detection:

```python
# Select South Asia data for 2022
sa_2022 = south_asia_data[south_asia_data['year'] == 2022].copy()

# HDI Outlier Limits
Q1_hdi = sa_2022['hdi'].quantile(0.25)
Q3_hdi = sa_2022['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

hdi_low = Q1_hdi - 1.5 * IQR_hdi
hdi_high = Q3_hdi + 1.5 * IQR_hdi

# GNI Outlier Limits
Q1_gni = sa_2022['gross_inc_percap'].quantile(0.25)
Q3_gni = sa_2022['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

gni_low = Q1_gni - 1.5 * IQR_gni
gni_high = Q3_gni + 1.5 * IQR_gni

# Find outliers
outliers = sa_2022[
    (sa_2022['hdi'] < hdi_low) | (sa_2022['hdi'] > hdi_high) |
    (sa_2022['gross_inc_percap'] < gni_low) | (sa_2022['gross_inc_percap'] > gni_high)
]

# Normal countries
normal = sa_2022.drop(outliers.index)

# Plot
plt.figure(figsize=(10, 6))

plt.scatter(normal['gross_inc_percap'], normal['hdi'], label='Normal', alpha=0.6)
plt.scatter(outliers['gross_inc_percap'], outliers['hdi'], label='Outliers', s=150)

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('Outliers: GNI vs HDI (South Asia 2022)')
plt.legend()
plt.grid(True)

plt.show()

```

4. Exploring Metric Relationships:

```python
# Clean South Asia data
sa = south_asia_data[['hdi', 'gender_development', 'life_expectancy']].dropna()

# Correlations
corr_gender = sa['gender_development'].corr(sa['hdi'])
corr_life = sa['life_expectancy'].corr(sa['hdi'])

# Create plots
plt.figure(figsize=(14, 5))

# Plot 1: Gender Development vs HDI
plt.subplot(1, 2, 1)
sns.regplot(data=sa, x='gender_development', y='hdi')
plt.title(f'Gender Development vs HDI\nCorrelation: {corr_gender:.4f}')
plt.xlabel('Gender Development Index')
plt.ylabel('HDI')

# Plot 2: Life Expectancy vs HDI
plt.subplot(1, 2, 2)
sns.regplot(data=sa, x='life_expectancy', y='hdi')
plt.title(f'Life Expectancy vs HDI\nCorrelation: {corr_life:.4f}')
plt.xlabel('Life Expectancy (years)')
plt.ylabel('HDI')

plt.tight_layout()
plt.show()
```

5. Gap Analysis:

```python
# Create GNI–HDI gap
dataset['GNI_HDI_Gap'] = dataset['gross_inc_percap'] - dataset['hdi']

# South Asian countries
sa_list = [
    'Afghanistan', 'Bangladesh', 'Bhutan', 'India',
    'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka'
]

# Filter South Asia data
sa = dataset[dataset['country'].isin(sa_list)]

# Get latest year for each country
sa_latest = sa.sort_values('year').groupby('country').tail(1)

# Top 3 positive and negative gaps
top_pos = sa_latest.sort_values('GNI_HDI_Gap', ascending=False).head(3)
top_neg = sa_latest.sort_values('GNI_HDI_Gap').head(3)

# Combine
gap_data = pd.concat([top_pos, top_neg])

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=gap_data, x='GNI_HDI_Gap', y='country')

plt.title('Top GNI–HDI Gaps in South Asia')
plt.xlabel('GNI – HDI Gap')
plt.ylabel('Country')

plt.show()

```

```python
# HDI Outlier calculation for South Asia
Q1_hdi = south_asia['hdi'].quantile(0.25)
Q3_hdi = south_asia['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi
hdi_lower = Q1_hdi - 1.5 * IQR_hdi
hdi_upper = Q3_hdi + 1.5 * IQR_hdi

# GNI Outlier calculation for South Asia
Q1_gni = south_asia['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni
gni_lower = Q1_gni - 1.5 * IQR_gni
gni_upper = Q3_gni + 1.5 * IQR_gni

# Detect outliers
outliers_SA = south_asia[
    (south_asia['hdi'] < hdi_lower) | (south_asia['hdi'] > hdi_upper) |
    (south_asia['gross_inc_percap'] < gni_lower) | (south_asia['gross_inc_percap'] > gni_upper)
]

print("South Asia Outliers:")
print(outliers_SA[['country', 'hdi', 'gross_inc_percap']])

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(south_asia['gross_inc_percap'], south_asia['hdi'], label='Normal Countries')
plt.scatter(outliers_SA['gross_inc_percap'], outliers_SA['hdi'], color='red', label='Outliers')
plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('South Asia: Outlier Detection (HDI vs GNI per Capita)')
plt.legend()
plt.show()
```
